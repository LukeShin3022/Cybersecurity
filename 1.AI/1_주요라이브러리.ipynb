### Aggregation (집계)

```python
# aggregation 기능은 데이터 탐색 단계에서 기초 통계량을 추출할 때 사용된다.
# 각 요소의 합
mat.sum()
```

### Boolean array 활용법

```python
# 비교
mat>5
mat[mat>5]
```

```python
# 비교 (논리 연산자 복합적 사용)
(mat > 5)&(mat < 10)
```

```python
# Boolean Array를 이용한 element 선택 - 1
mat[mat>5]
```

```python
# Boolean Array를 이용한 element 선택 - 2
mat[(mat > 5)&(mat < 10)]
```

```python
# Boolean Array를 이용한 element 선택 후 aggregation(집계) - 1
np.sum(mat>5)
```

```python
# Boolean Array를 이용한 element 선택 후 aggregation(집계) - 2
np.sum(mat[mat>5])
```

# Matplotlib 라이브러리

```python
#라이브러리 로드
import numpy as np
import matplotlib.pyplot as plt

# plotting하기 위한 샘플데이터 생성
X = np.linspace(0, 100, 20) # 0~100까지 균일하게 20point
Y1 = X
Y2 = X**2
```

```python
x = np.linspace(0, 100, 10)
y = x**2
plt.plot(x,y,label='x*2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('AI python class')
plt.legend()
plt.show()
```

```python
plt.plot(X, Y2, label='a')
plt.legend()
```

```python
# stateless 하게 plotting
fig, axes = plt.subplots(2,1, figsize=(10,5))
axes[0].plot(X,Y1)
axes[1].plot(X,Y2)
```

```python
# stateful 하게 plotting
plt.figure(figsize=(10,5))
plt.subplot(2,1,1)
plt.plot(X,Y1)
plt.subplot(2,1,2)
plt.plot(X,Y2)
plt.show()
```
