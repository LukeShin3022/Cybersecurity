```python
### Tuple Quiz ## 속도가 List에 비해 빠르다
count1 = []
for i in range (7):
  count1.append(i)

count_tuple = tuple(count1)

count_list = list(count_tuple)


count_list[2] = 7
count_tuple = tuple(count_list)
print(count_tuple)
print(count_list)

myTuple = ('tuple',(1,2),[3,4]) # Tuple 안에 List 혹은 Tuple을 추가로 중첩할 수 있다.
print(myTuple[-1][0])
myTuple[-1][0] = 7 # Tuple은 바뀌지 않지만 Tuple 내부의 List 혹은 변경가능한 구조는 바꿀 수 있다.
for i in myTuple:
  print(i)

print(myTuple)

```

```python
### Upper case
myList = ['one','two','three']
# myList1 = myList ## 대입이 아니라 참조가 된다 그래서 myList == myList1이 되어 버림
myList1 = list(myList) ## 대입이 아니고 Copy 하고 싶다면 list 함수를 사용: List객체를 생성하는 방법
myList2 = list(myList)

for i in range(3):
  myList1[i] = myList1[i].upper()
  print(myList1[i])

# print(myList2[0][0])

for i in range(3):
  myList2[i] = myList2[i].capitalize()
  print('\n', myList2[i])

for i in range(3):
  myList2[i] = myList2[i][0].upper() + myList2[i][1:]
  print('\n', myList2[i])
```

```python
### 중첩 for문을 이용하자 그리고 함수로 만들어 보자
num = 0
num2 = 0
num3 = 21

def increseNum(num):
  return print(f'{num}', end='\t')

def timesNum(num2):
  return print(f'{num2 * 5}', end='\t')

def decreaseNum(num3):
  return print(f'{num3}', end='\t')


for i in range(5):
  for j in range(12):

    if j < 4:
      num += 1
      # print(f'{num}', end='\t')
      increseNum(num)

    elif j < 8:
      num2 += 1
      timesNum(num2)
      # print(f'{num2 * 5}', end='\t')

    elif j < 12:
      num3 -= 1
      decreaseNum(num3)
      # print(f'{num3}', end='\t')

  print('\n')

  ### Operation

def add(a,b):
  return a + b

def sub(a,b):
  return a - b

print(add(1,2))
print(sub(1,2))
```

```python
### Calculator

num1 = input('첫번째 숫자를 입력하세요: ')
num2 = input('두번째 숫자를 입력하세요: ')
ope = input('연산자를 입력하세요(e.g +, -, *, /, abs): ')

num1 = float(num1)
num2 = float(num2)

if ope == '+':
  print(f'{num1} + {num2} = {num1 + num2}')

elif ope == '-':
  print(f'{num1} - {num2} = {num1 + num2}')

elif ope == '*':
  print(f'{num1} * {num2} = {num1 + num2}')

elif ope == '/':
  print(f'{num1} / {num2} = {num1 + num2}')

elif ope == 'abs':
  num1 = input('절대값 계산을 위해 숫자를 다시 입력하세요: ')
  num1 = float(num1)
  if num1 >=0:
    print(f'절대값은 {num1} 입니다')
  else:
    print(f'절대값은 {-num1} 입니다')

else:
  print('연산자를 다시 입력하세요')

```
