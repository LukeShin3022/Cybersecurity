```python
### Baseball game

# 야구게임은 입력받은 숫자의 위치의 일치여부와 숫자의 일치 여부를 가지고 3OUT을 잡아내는 게임이다.
# 자리와 숫자가 맞으면 1 strike, 숫자만 맞으면 1ball을 판단한다.
# 먼저 Strike를 판단하고 그다음에 Ball을 판단해보면 될까?
# 정답을 구하는 건 Random 수를 List에 넣는다. 각자리의 수는 중복될 수 없게 만든다## Baseball game
# 야구게임은 입력받은 숫자의 위치의 일치여부와 숫자의 일치 여부를 가지고 3OUT을 잡아내는 게임이다.
# 자리와 숫자가 맞으면 1 strike, 숫자만 맞으면 1ball을 판단한다.
# 먼저 Strike를 판단하고 그다음에 Ball을 판단해보면 될까?
# 정답을 구하는 건 Random 수를 List에 넣는다. 각자리의 수는 중복될 수 없게 만든다
# 정답이 123일 때, 자리수에 숫자가 맞는지를 확인하고 strike를 판단한다
# 생각해 보니 Ball을 먼저 판단하는게 좋겠다.
# Ball이 3개이면 Strike가 섞여 있고 Strike를 판단할 때, cnt해서 Ball에서 빼면될거 같다.
# Strike를 판단 시 만약 strike==3, Ball==3이면 루프를 나오고 정답 처리를 한다.

# 1. 1-9의 랜덤 수를 넣은 List를 생성한다.
# 2. 정답을 랜덤 수에서 추출해서 넣는다
# 3. ball의 여부를 판단한다.
# 4. Strike 여부를 판단한다.

import random as rd
import sys

def genRanNum():
  ran_num = []
  for i in range (1,10):
    ran_num.append(i)
  rd.shuffle(ran_num)
  print(ran_num)
  return ran_num

def selOutNum(ran_numList):
  out_num = []
  for j in range(3):
    out_num.append(ran_numList.pop())
  print(out_num)
  return out_num

def chkBall(num1, out_num):
  ballCnt = 0
  # print('ball',num1)
  for k in range(3):
    # print(num1[k])
    if num1[k] in out_num:
      ballCnt += 1
  # print(f'ball: {ballCnt}')
  return ballCnt

def chkStrike(num1, out_num):
  strikeCnt = 0
  for h in range(3):
    if num1[h] == out_num[h]:
      strikeCnt += 1
  # print(f'strikeCnt: {strikeCnt}')
  return strikeCnt

def inpNum():
  inp_Num = []
  for l in range(3):
    if 99 in inp_Num:
      print('게임을 종료합니다.')
      sys.exit()
    inp_Num.append(int(input(f'3개의 숫자 입력, {l+1}번째 숫자를 입력하세요 ')))
  # print(inp_Num)
  return inp_Num

out_num = selOutNum(genRanNum())

while True:
  ball_cnt = 0
  str_cnt = 0

  inp_Num = inpNum()
  ball_cnt = chkBall(inp_Num, out_num)
  str_cnt = chkStrike(inp_Num, out_num)

  if str_cnt == 3 and ball_cnt == 3:
    print(f'Strike OUT!. 입력값:{inp_Num}, 정답:{out_num}')
    break

  else:
    if str_cnt > 0:
      ball_cnt = ball_cnt - str_cnt
      print(f'Strike: {str_cnt} Ball: {ball_cnt}')

    elif str_cnt < 1:
      print(f'Strike: {str_cnt} Ball: {ball_cnt}')
```

```python
### Set(집합) # 중복값이 없는 자료형
dic = {}
print(type(dic))

tSet = set(dic)
print(type(tSet)) # convert dictionary to set

tDic = dict(tSet) # convert set to dictionary
print(type(tDic))

mySet = {1,2,3}
print(mySet)
print(type(mySet))

mySet1 = {1,2,3,3,4,2}
mySet2 = {1,4,6,3,4,2,6,7,10,11}

print(mySet)

### 집합에 항목 추가
mySet = {'red', 'blue'}

mySet.add('yellow')
print(mySet)

mySet.update({1,2},[2,3])
print(mySet)

# 집합은 논리 연산을 할 수 있다. 합집합, 교집합, 차집합 등등

print(mySet1 | mySet2) # Union
print(mySet1 & mySet2) # Intersection
print(mySet1 - mySet2) # difference
print(mySet1 ^ mySet2) # symetric_difference
```

```python
### Set Quiz
### List안에 중복 단어들의 수를 카운드 하는 프로그램

word = ['apple','bananA','Apple','bAnana','cherRy','chErry','apPle','Strawberry','StrawberRy','OrAnge']
uni_word = [i.lower() for i in word]

### Set()을 사용
# print(uni_word)
for item in set(uni_word):
    cnt = uni_word.count(item)
    print(f"{item}: {cnt}")


print()
### Set을 사용하지 않는 방법
def remDup():
  word2 = []
  for i in range(len(uni_word)):
    if uni_word[i] not in word2:
      word2.append(uni_word[i])
  return word2

word2 = remDup()
# print(word2)

for item in word2:
  cnt = uni_word.count(item)
  print(f"{item}: {cnt}")

print()
### Dictionary를 이용하는 방법
wDic = {}
for item in uni_word:
  if item in wDic.keys():
    wDic[item] += 1
  else:
    wDic[item] = 1
print(wDic)


print()
### 노가다로 찾는 방법
for i in range(len(word2)):
  cnt = 0
  for j in range(len(word)):
    if word2[i] == word[j].lower():
      cnt += 1
  print(f'{word2[i]}: {cnt}')

```
