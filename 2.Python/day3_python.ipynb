```python
### 문장에서 the, that같은 용어를 제거하자

# 문장에서 제거하기 원하는 단어를 받아서 원문에서 제거하기
# 1. 문장을 split해서 List로 만들기
# 2. 단어 받아서 List로 만들기
# 3. List의 length까지 돌아서 문장List에서 일치하는 단어 제거하기
# 4. 다시 문장으로 만들어 주기

sentence = "The regret after not doing something is bigger than that of doing something"
rem_words = []

class EditSentence:
  def __init__(self, sentence):
    self.sentence = sentence

  def splitSentence(self, sperator):
    return self.sentence.split(sperator)

  def removeSentence(self, word, list):
    for w in word:
      if w in list:
        list.remove(w)
    return list

  def joinList(self, list):
    return ' '.join(list)


while True:
  rem_word = input('문장에서 삭제하길 원하는 단어를 입력하시오(종료는 x): ')
  if rem_word == 'x':
    break
  else:
    rem_words.append(rem_word)

es = EditSentence(sentence)
result = es.joinList(es.removeSentence(rem_words, es.splitSentence(' ')))
print(result)
```

```python
### 예외처리 Quiz
# 단어를 입력하면 해당 단어가 파일에 존재하지 않으면 추가하는 프로그램을 만들기
# 1.단어 입력 받기 List로
# 2.파일 입출력 함수 만들기
# 3.파일 Search 함수 만들기
# 4.다시생각 하자! 기능을 구분하자 읽는 함수, 찾는 함수, 다시 파일에 쓰는 함수
# 5.파일객체는 구분하면 안된다 복잡해진다
# 6.기존에 파일에 있는 값은 추가가 되지 않지만 파일에 없는 단어를 한번에 여러개 입력하게 되면 파일에 write가 된다 이 문제를 해결하기 위해 입력 받은 단어를 List가 아니라 Set에 넣었다
from google.colab import drive
drive.mount('/content/drive')
drivePath = '/content/drive/MyDrive/KISIA Class for Cybersecurity/2.Python/' # 역슬래쉬를 뺴야한다

def fileWrite(words):
  with open(drivePath + 'addWord.txt', 'a+') as f:
    f.seek(0)
    fileBody = f.read().splitlines()
    for word in words:
      if word not in fileBody:
        f.write('\n' + word)
        print(f"신규단어가 추가됨: {word}")
      else:
        print(f"이미 존재하는 단어: {word}")
    # print(f.seek(0),f.read())

words = set()
while True:
  word = input('파일에 추가할 단어를 입력하세요(입력완료 x): ')
  if word == 'x':
    break
  else:
    words.add(word)

fileWrite(words)
```

```python
### Pandas 문자열 처리
import pandas as pd

df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'CharLie', 'DAvE'],
    'city': ['SeOul', 'BusAn', 'DaeGu', 'inChon']
})

# 대소문자 변환
df['name_lower'] = df['name'].str.lower()
df['name_upper'] = df['name'].str.upper()

# 공백제거
df['name_strip'] = df['name'].str.strip()
df['city_strip'] = df['city'].str.strip()

# 특정 문자 대체
df['name_replace'] = df['name_strip'].str.replace('a', '@', case=False)

# 문자열 길이 및 포함 확인
df['name_length'] = df['name_strip'].str.len()
df['has_char_l'] = df['name_strip'].str.contains('l', case=False)

print(df)
```

```python
### pandas 데이터 정렬

import pandas as pd

df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charle', 'David'],
    'score': [85,92,78,90],
    'join_date': ['2025-01-10', '2026-3-15', '2027-07-09', '2025-01-30']
})

# 날짜형으로 변환
df['join_date'] = pd.to_datetime(df['join_date'])

# 점수 기준 내림차순 정렬

sorted_df = df.sort_values(by='score', ascending=False)
print(sorted_df)
```

```python
### Pandas Data filtering

# 점수가 85 이상인 데이터만 추출
high_score = df[df['score'] >= 85]
print(high_score)

print()
# join_date가 2025년 2월 이후인 데이터
filtered = df.query("join_date >= '2025-02-01'")
print(filtered)
```

```python
### Data merge

students = pd.DataFrame({
    'id': [1,2,3],
    'name': ['Alice', 'Bob', 'Charlie']
})

score = pd.DataFrame({
    'id': [1,2,4],
    'score': [85,92,78]
})

merged_df = pd.merge(students, score, on='id', how='inner')
print(merged_df)
print()

merged_outer = pd.merge(students, score, on='id', how='outer')
print(merged_outer)
print()

merged_left = pd.merge(students, score, on='id', how='left')
print(merged_left)
print()

merged_right = pd.merge(students, score, on='id', how='right')
print(merged_right)

## 두 데이터를 수직으로 병합
df1 = pd.DataFrame({'name': ['Alice', 'Bob'],'score': [85, 90]})
df2 = pd.DataFrame({'name': ['Charlie', 'David'],'score': [78,92]})

print(df1)
print()
print(df2)
print()

concat_df = pd.concat([df1, df2], ignore_index=True) # False인 경우 index를 포함한 채로 합쳐진다 즉 index가 중복될 가능성이 있음
print(concat_df)
```
